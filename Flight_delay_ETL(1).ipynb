```python
import pandas as pd
df=pd.read_csv("civil_aviation.csv")
print(df.info())
print(df.head())
```

```python
df.rename(columns={"Month":"month","Number of Departures":"flights_departed","Number of flights Delayed":"flights_delayed","% On-time Performance":"Ontime_percentage"})
```

```python
df
```

```python
df['Month'] = df['Month'].astype(str).str.strip()
```

```python
df['Month'] = df['Month'].str.replace('Auguest 2024', 'Aug-24')
```

```python
df['Month'].unique()
```

```python
#conversion to datetime format
df['Month'] = pd.to_datetime(df['Month'], format='%b-%y', errors='coerce')
```

```python

# Check for NaT values
print(df[df['Month'].isna()])  # Debugging step to find issues

# Save cleaned file
df.to_csv("cleaned_file.csv", index=False)
```

```python
import pandas as pd
df2=pd.read_csv("cleaned_file.csv")
print(df2.info())
print(df2.head())
```

```python
df2['Number of Departures'] = df2['Number of Departures'].astype(int)  # Convert to int
df2['Number of flights Delayed'] = df2['Number of flights Delayed'].astype(int)  # Convert to int
#df2['% On-time Performance'] = df2['% On-time Performance'].str.replace('%', '').astype(float)  # Convert to float
df2
```

```python
df2['% On-time Performance']=df2['% On-time Performance'].astype(str)
df2['% On-time Performance']=df2['% On-time Performance'].str.replace(" ","")
df2['% On-time Performance']=df2['% On-time Performance'].astype(str)
print(df2['% On-time Performance'].unique())
```

```python
df2['calculated_performance2']=((df2['Number of Departures']-df2['Number of flights Delayed'])/df2['Number of Departures'])*100
```

```python
print(df2['% On-time Performance'].unique())
print('--------------------------------------')
print(df2['calculated_performance2'].unique())
```

```python
df2.to_csv("cleaned_file.csv", index=False)
```

```python
df2['% On-time Performance'] = pd.to_numeric(df2['% On-time Performance'], errors='coerce')
threshold=0.1
mask=abs(df2['% On-time Performance']-df2['calculated_performance2'])> threshold
df2.loc[mask,'% On-time Performance']=df2['calculated_performance2']
df2.drop(columns=['calculated_performance2'],inplace=True)
df2.to_csv("cleaned_file.csv",index=False)
```

```python
df2['% On-time Performance']=df2['% On-time Performance'].round(2)
df2['calculated_performance']=df2['calculated_performance'].round(2)
df2.to_csv("cleaned_file.csv",index=False)
```

```python
df2=df2.rename(columns={'calculated_performance':'Delay Rate','% On-time Performance':'On time Arrival Rate'})
df2.to_csv("cleaned_file.csv",index=False)
```
